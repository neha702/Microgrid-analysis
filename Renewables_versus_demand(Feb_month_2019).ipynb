```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**1 Feb 2019**

```
#Data preprocessing
demand_data=pd.read_csv('/content/CAISO-demand-20190201.csv')
supply_data=pd.read_csv('/content/CAISO-supply-20190201.csv')
X=demand_data.iloc[2,:-1].values
Y=supply_data.iloc[0,:-1].values
df1=pd.DataFrame(X)
headers=df1.iloc[0]
df1=pd.DataFrame(df1.values[1:],columns=headers)
df2=pd.DataFrame(Y)
headers2=df2.iloc[0]
df2=pd.DataFrame(df2.values[1:],columns=headers2)
print(df1)
print(df2)
print(df2.shape)
#Check for NaN columns
df1.isnull().sum()
#df1.fillna(method='ffill',axis=0,inplace=True)
data=df1.join(df2,how="inner")
print(data.head(5))
#Jointplot and correlation
df1=data["Demand (5 min. avg.)"].astype(float)
df2=data["Renewables"].astype(float)
print(df1.corr(df2))

sns.jointplot(x = "Demand (5 min. avg.)", y = "Renewables", 
              kind = "scatter", data = data)
fig,ax=plt.subplots(figsize=(20,5),ncols=2,nrows=1,squeeze=False)
#1d kde plot
data=data.astype(float)
sns.kdeplot(data=data,ax=ax[0][0])
#2d kde plot
df1=data["Renewables"].astype(float)
df2=data["Demand (5 min. avg.)"].astype(float)
sns.kdeplot(data=data,x=df2,y=df1,color='r',shade=True,shade_lowest=False,cmap="Reds",ax=ax[0][1])
```
