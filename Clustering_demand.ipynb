```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Data preprocessing
demand_data=pd.read_csv('/content/drive/MyDrive/Battery analysis/Battery analysis/CAISO-demand-20190201.csv')
demand_data=demand_data.transpose()
headers=demand_data.iloc[0]
df1=pd.DataFrame(demand_data.values[1:],columns=headers)
time1=[]
for i in range(0,1441,5):
    time1.append(i)                 #making up an array of time axis and inserting values using append
df1["Time(in min)"]=pd.DataFrame(time1)
df1=df1[["Time(in min)","Demand (5 min. avg.)"]]
print(df1.head())
print(df1.isnull().sum())
df1=df1.fillna(df1.mean())
print(df1.isnull().sum())
import scipy as sp
import scipy.fftpack
from numpy.fft import fft,fftfreq
n=289
Lx=1440
xx=np.linspace(0,Lx,n)
x=df1["Demand (5 min. avg.)"] 
fft1= np.fft.fft(x)   #Raw fft values
fftabs=np.abs(fft1) #Processed absolute fft values
fftang=np.angle(fft1) 
freq=fftfreq(n,1/(5*60))
mask=freq>0
fig, ax = plt.subplots(4, 1, figsize=(10, 8))
ax[0].plot(xx,x)
ax[0].set_title('Original Signal')
ax[1].plot(freq[mask],(fftabs[mask]))
ax[1].set_title('Magnitude spectrum')
ax[2].plot(freq[mask],(fftang[mask]))
ax[2].set_title('Phase spectrum')
sns.kdeplot(df1['Demand (5 min. avg.)'].astype(float),ax=ax[3]) #Time domain kde plot
#sns.kdeplot(fftabs,ax=ax[4]) #Magnitude spectrum kde plot
plt.tight_layout() #Plot titles and axes titles do not overlap. 
demand1=df1["Demand (5 min. avg.)"].to_numpy()
#list.append(demand1)
```
